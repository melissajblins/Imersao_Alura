Montar gráfico com o seaborn:

Para o caso da coluna “droga”, não é viável fazer um gráfico de todos os elementos, uma vez que ficaria impossível de ser lido e interpretado. **Solução: Exibir os 5 elementos mais citados**. 

Código para todos os elementos:

```
import seaborn as sns

sns.countplot(x = 'droga', data = dados)
```

````python
maiores_compostos = dados['droga'].value_counts().index[0:5]
maiores_compostos
````

````python
dados.query('droga in @maiores_compostos')
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'droga', data = dados.query('droga in @maiores_compostos'))
ax.set_title('5 compostos mais utilizados')
plt.show()
````
